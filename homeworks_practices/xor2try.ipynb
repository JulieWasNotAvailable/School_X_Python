```python
from keras import models
from keras import layers
from keras import optimizers
```

```python
import numpy as np
```

```python
#данные для обучения
x = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
y = np.array([[1,0],[0,1],[0,1],[1,0]], dtype=float)
#в x - смотрим на второе значение, первое - обратное
```

Обоснование к тому, почему мы должны использовать два слоя: <br>
С помощью одного слоя нейросети мы способны построить на координатной прямой линию (или даже две), задав коэффициент (k) и смещение, bias (kx+b). Однако задача не решается одной линией. <br>
Когда мы создаём новый слой (часть справа на картинке), все возможные выходы узлов становятся координатами (0,0), (0,1) и (2,1), которые уже можно разбить на подгруппы одной прямой


```python
opt = optimizers.SGD(learning_rate = 1)
#градиентный стохастический спуск со скоростью обучения - 1
```

```python
model = models.Sequential()
model.add(layers.Dense(units = 4, activation='sigmoid'))
model.add(layers.Dense(units = 2, activation='sigmoid')) #relu не работает
```

```python
model.compile(loss="categorical_crossentropy", optimizer=opt, metrics="accuracy")
```

```python
model.fit(x,y, epochs=500)
```

```python
pred = model.predict_on_batch(x)
pred
```
