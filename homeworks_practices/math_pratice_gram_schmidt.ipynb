```python
class Matrix():
  
    def __init__(self, matrix: list):
        self.matrix = matrix

    def T(self):

        m1 = self.matrix

        n_rows1 = len(m1)
        n_cols1 = len(m1[0]) 

        m2 = []
        for n in range(0,n_cols1):
            m2 = m2 + [[0]*n_rows1]

        r = 0
        c = 0
        for row in m2:
            for el in row:
                m2[r][c] = m1[c][r]
                c+=1
            c = 0
            r+=1
        
        m1 = Matrix(m2)
        return (m1)            

    def __add__(self, n: int|float|list):

        if (type(n).__name__ == "int") or (type(n).__name__ == "float"):

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m2 = []
            for num in range(0,n_rows):
                m2 = m2 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m2:
                for el in row:
                    m2[r][c] = self.matrix[r][c] + n
                    c += 1
                c = 0
                r+=1

            m1 = Matrix(m2)
            return (m1)  

        else:

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for num in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = self.matrix[r][c] + n.matrix[r][c]
                    c += 1
                c = 0
                r+=1
    
            m1 = Matrix(m3)
            return (m1)  
    
    def __sub__(self, n: int|float|list):

        if (type(n).__name__ == "int") or (type(n).__name__ == "float"):       

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for num in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = self.matrix[r][c] - n
                    c += 1
                c = 0
                r+=1

            m1 = Matrix(m3)
            return (m1) 
        
        else: 

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for num in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = self.matrix[r][c] - n.matrix[r][c]
                    c += 1
                c = 0
                r+=1 
            
            m1 = Matrix(m3)
            return (m1)            
    
    def __truediv__(self, n:int|float|list):

        if (type(n).__name__ == "int") or (type(n).__name__ == "float"):    

            m1 = self.matrix      

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for num in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m1:
                for el in row:
                    m3[r][c] = m1[r][c]/n
                    c += 1
                c = 0
                r+=1

            m2 = Matrix(m3)
            return (m2) 
        
        else:

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for smth in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = self.matrix[r][c] / n.matrix[r][c]
                    c += 1
                c = 0
                r+=1
    
            m1 = Matrix(m3)
            return (m1)                  

    def __mul__(self, n: int|float|list): #название аргумента функции - n, потому что неизвестно, инт или лист

        if (type(n).__name__ == "int") or (type(n).__name__ == "float"):

            m1 = self.matrix   

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for smth in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]   

            r = 0 #row
            c = 0 #column
            for row in m1:
                for el in row:
                    m3[r][c] = m1[r][c]*n
                    c += 1
                c = 0
                r+=1

            m2 = Matrix(m3)
            return (m2) 
        
        else:

            m1 = self.matrix
            m2 = n.matrix

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for num in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]
            
            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = m1[r][c] * m2[r][c]
                    c += 1
                c = 0
                r+=1

            m1 = Matrix(m3)
            return (m1) 

    def __matmul__(self, m: list):

            m1 = self.matrix
            m2 = m.matrix

            n_rows1 = len(m1)
            n_cols2 = len(m2[0]) 

            m3 = []
            for n in range(0,n_rows1):
                m3 = m3 + [[0]*n_cols2]
            
            for i in range(len(m1)):
                for j in range(len(m2[0])):
                    for k in range(len(m2)):
                        m3[i][j] += m1[i][k] * m2[k][j]

            m1 = Matrix(m3)
            return (m1)
```

```python
class Vector(Matrix):

    def gram_schmidt(v1,v2,v3,v4: list):

        def vlen(vector: list): #finds length of the vector (returns scalar)

            vector = vector.matrix

            len = 0
            for row in vector:
                for el in row:
                    len = len + el*el

            return (len**0.5)

        def vmul(self, v:list): #multiplication of vectors (returns scalar)

            v1 = self.matrix
            v2 = v.matrix
            
            r = 0 #row
            c = 0 #column
            prod = 0 #product
            for row in v1:
                for el in row:
                    prod = prod +v1[r][c] * v2[r][c]
                    c += 1
                c = 0
                r+=1

            return (prod)

        def null_counter(self): #находит нули, если четыре, то вектор нулевой, соответственно его не берём
            
            vector = self.matrix

            counter = 0
            for row in vector:
                for el in row:
                    if el == 0:
                        counter+=1
            
        u1 = v1
        e1 = u1/vlen(u1)

        u2 = v2 - u1*(vmul(u1,v2)/vmul(u1,u1))
        e2 = u2/vlen(u2)

        u3 = v3 - u1*(vmul(u1,v3)/vmul(u1,u1)) - u2*(vmul(u2,v3)/vmul(u2,u2))
        e3 = u3/vlen(u3)

        u4 = v4 - u1*(vmul(u1,v4)/vmul(u1,u1)) - u2*(vmul(u2,v4)/vmul(u2,u2)) - u3*(vmul(u3,v4)/vmul(u3,u3))
        e4 = u4/vlen(u4)

        basis = [e1, e2, e3, e4]

        for vector in basis:
            for row in vector.matrix:
                for num in row:
                    num = round(num,2)

        basisv = [x.matrix for x in basis if null_counter(x)!=4]
        
        basisv = Matrix(basisv)
        return (basisv)
```
