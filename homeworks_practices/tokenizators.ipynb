```python
class MyTextTokenizer:
    
    def __init__(self, adress):
        self.text = open(adress, "r")
        self.text_list = self.text.read()

    def my_tokenizer(self, text):

        text_tok = text.split(" ")

        good_text_tok = []
        for token in text_tok:
            good_text_tok.append(re.sub(r'\n', r"", token))

        for el in good_text_tok: #i - element
            if el == "":
                good_text_tok.remove(el)

        return good_text_tok #good_text_tok
```

```python
pooh_book = MyTextTokenizer("datasets\winnie_pooh.txt")
pb = pooh_book.my_tokenizer(pooh_book.text_list)

pb
```
