```python
from nltk import WhitespaceTokenizer
from nltk.corpus import stopwords
```

```python
#получить список самых часто повторяющихся слов, прилоэить к ним какие-нибудь векторы, посмотреть, что будет
stop = stopwords.words("russian")
```

```python
def text_preproc(adress):
    text = open(adress, "r", encoding = "utf-8").read()
    text = WhitespaceTokenizer().tokenize(text)
    words_filtered = [word for word in text if word not in stop]

    new = []
    for word in words_filtered:
        if "." in word:
            word = word.replace(".", "")
            if "." in word:
                word = word.replace(".", "")
                if "." in word:
                    word = word.replace(".", "")
        if "," in word:
            word = word[0:len(word)-1]
        if "\"" in word or "\n" in word:
            word = word.replace("\"", "")
            word = word.replace("\n", "")
            
        new.append(word)
    return new
```

```python
text = text_preproc("../homeworks_practices/6_texts/antonovskie-yabloki.txt")

for element in text:
    if "I" in element:
        text.remove(element)
```
