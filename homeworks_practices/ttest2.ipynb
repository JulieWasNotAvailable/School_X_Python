1 Задание: <br>
Приведены два датасета: время использования одними и теми же пользователями личного кабинета сайта до редизайна и после. Проверьте гипотезу, что после редизайна время использования личного кабинета на сайте изменилось (увеличилось или уменьшилось). Установите уровень статистической значимости 0.05. <br>

(подсказка: выборки зависимые друг от друга, так как тест проводится над одними и теми же пользователями, поэтом найдите подходящий для таких проверок ttest у scipy) <br>

$H_0$: Время использования не изменилось после редизайна (до=после) <br>
$H_1$: Время использования после редизайна изменилось

```python
import scipy.stats as st
```

```python
time_before = [1732, 1301, 1540, 2247, 1632, 1550, 754, 1946, 1889,
          2748, 1349, 1648, 1665, 2416, 1470, 1681, 1868, 1629,
          1271, 1633, 2131, 942, 1599, 1127, 2200, 661, 1207,
          1737, 2410, 1486]

time_after = [955, 2577, 360, 139, 1618, 990, 644, 1796, 1487, 949, 472,
         1906, 1758, 1258, 2554, 612, 309, 1864, 1294, 1487, 1164, 1559,
         491, 2286, 1270, 2069, 1553, 1629, 1704, 1623]
```

```python
alpha = 0.05
```

```python
results = st.ttest_rel(
    time_before,
    time_after,
    alternative = "two-sided"
)

if results.pvalue < alpha:
    print ("Отвергаем нулевую гипотезу")
else:
    print("Нет оснований отвергнуть нулевую гипотезу")
```

2 Задание: <br>

У вас есть два датасета: в них количество патронов, которые покупают постоянные игроки в онлайн-игре, до и после изменения. Проверьте гипотезу о том, что после нововведений количество используемых патронов изменилось. <br>

$H_0$: Количество используемых патроновф не изменилось (до=после) <br>
$H_1$: Количество используемых патронов изменилось (до!=после)

```python
results = st.ttest_rel(
    bullets_before,
    bullets_after,
    alternative = "two-sided"
)

if results.pvalue < alpha:
    print ("Отвергаем нулевую гипотезу")
else:
    print("Нет оснований отвергнуть нулевую гипотезу")
```
