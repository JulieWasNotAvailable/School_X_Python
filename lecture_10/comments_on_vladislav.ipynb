```python
df = pd.read_csv("python_task_dataset.csv")
#понять, какие пользователи чаще всего кликают на рекламу, назвать их характеристики

# Device_id – уникальный идентификатор девайса отправителя события
# Device_type – тип девайса android, ios
# Datetime – время отправки события - во сколько он кликнул или посмотрел
# Ev_type – тип события (0 – показ, 1 – клик) - только посмотрел или кликнул
# Adv_type – тип рекламы multi - всплывающее окно, видео - вставлялось в вк видео
# Region_id – регион отправителя события
# Age – возраст отправителя события 
# Gender – пол отправителя события (1 – женщины, 2 – мужчины)

#какие пользователи кликают в сравнении с теми, кто не кликают
#можно для каждого признака сформировать

clicked = df[df["ev_type"] == 1]
clicked = clicked.groupby("gender")
clicked.sum() #хочу посмотреть на график юзеров, которые кликали
```

```python
sns.barplot(clicked, x="gender", y = "", hue = "gender")
#когда явно передаёшь x y не передавай series в data
```
