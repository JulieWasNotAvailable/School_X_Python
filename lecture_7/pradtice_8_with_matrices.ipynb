```python
#я знаю, что код можно было упростить, но я устала, и боюсь всё испортить

class Matrix:

    def __init__(self, matrix: list):
        self.matrix = matrix

    def T(self):

        m1 = self.matrix

        n_rows1 = len(m1)
        n_cols1 = len(m1[0]) 

        m2 = []
        for n in range(0,n_cols1):
            m2 = m2 + [[0]*n_rows1]

        r = 0
        c = 0
        for row in m2:
            for el in row:
                m2[r][c] = m1[c][r]
                c+=1
            c = 0
            r+=1
        
        m1 = Matrix(m2)
        return (m1)            

    def __add__(self, n: int|float|list):

        if (type(n).__name__ == "int") or (type(n).__name__ == "float"):

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m2 = []
            for num in range(0,n_rows):
                m2 = m2 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m2:
                for el in row:
                    m2[r][c] = self.matrix[r][c] + n
                    c += 1
                c = 0
                r+=1

            m1 = Matrix(m2)
            return (m1)  

        else:

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for num in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = self.matrix[r][c] + n.matrix[r][c]
                    c += 1
                c = 0
                r+=1
    
            m1 = Matrix(m3)
            return (m1)  
    
    def __sub__(self, n: int|float|list):

        if (type(n).__name__ == "int") or (type(n).__name__ == "float"):       

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for num in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = self.matrix[r][c] - n
                    c += 1
                c = 0
                r+=1

            m1 = Matrix(m3)
            return (m1) 
        
        else: 

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for num in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = self.matrix[r][c] - n.matrix[r][c]
                    c += 1
                c = 0
                r+=1 
            
            m1 = Matrix(m3)
            return (m1)            
    
    def __truediv__(self, n:int|float|list):

        if (type(n).__name__ == "int") or (type(n).__name__ == "float"):    

            m1 = self.matrix      

            r = 0 #row
            c = 0 #column
            for row in m1:
                for el in row:
                    m1[r][c] = m1[r][c]/n
                    c += 1
                c = 0
                r+=1

            m2 = Matrix(m1)
            return (m2) 
        
        else:

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for smth in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]

            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = self.matrix[r][c] / n.matrix[r][c]
                    c += 1
                c = 0
                r+=1
    
            m1 = Matrix(m3)
            return (m1)                  

    def __mul__(self, n: int|float|list): #название аргумента функции - n, потому что неизвестно, инт или лист

        if (type(n).__name__ == "int") or (type(n).__name__ == "float"):

            m1 = self.matrix      

            r = 0 #row
            c = 0 #column
            for row in m1:
                for el in row:
                    m1[r][c] = m1[r][c]*n
                    c += 1
                c = 0
                r+=1

            m2 = Matrix(m1)
            return (m2) 
        
        else:

            m1 = self.matrix
            m2 = n.matrix

            n_rows = len(self.matrix)
            n_cols = len(self.matrix[0]) 

            m3 = []
            for num in range(0,n_rows):
                m3 = m3 + [[0]*n_cols]
            
            r = 0 #row
            c = 0 #column
            for row in m3:
                for el in row:
                    m3[r][c] = m1[r][c] * m2[r][c]
                    c += 1
                c = 0
                r+=1

            m1 = Matrix(m3)
            return (m1) 

    def __matmul__(self, m: list):

            m1 = self.matrix
            m2 = m.matrix

            n_rows1 = len(m1)
            n_cols2 = len(m2[0]) 

            m3 = []
            for n in range(0,n_rows1):
                m3 = m3 + [[0]*n_cols2]
            
            for i in range(len(m1)):
                for j in range(len(m2[0])):
                    for k in range(len(m2)):
                        m3[i][j] += m1[i][k] * m2[k][j]

            m1 = Matrix(m3)
            return (m1) 
```

```python
%%ipytest

matrix1 = Matrix ([
        [1,7,3],
        [4,5,2],
        ])

matrix2 = Matrix ([
        [10,-2, 6],
        [-4, 5, 8],
        ])

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2]])
def test_matrix_add_matrix(matrix1: Matrix, matrix2: Matrix):
    assert (
        (np.array((matrix1 + matrix2).matrix) == (np.array(matrix1.matrix) + np.array(matrix2.matrix))).all()
    ), 'add section with matrices is not working right'

@pytest.mark.parametrize("matrix1,number", [[matrix1, 13.2]])
def test_matrix_add_number(matrix1: Matrix, number: int | float):
    assert (
        (np.array((matrix1 + number).matrix) == np.array(matrix1.matrix) + number).all()
    ), 'add section with matrix and number is not working right'

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2]])
def test_matrix_sub_matrix(matrix1: Matrix, matrix2: Matrix):
    assert (
        (np.array((matrix1 - matrix2).matrix) == np.array(matrix1.matrix) - np.array(matrix2.matrix)).all()
    ), 'sub section with matrices is not working right'

@pytest.mark.parametrize("matrix1,number", [[matrix1, 12.2]])    
def test_matrix_sub_number(matrix1: Matrix, number: int | float):
    assert (
        (np.array((matrix1 - number).matrix) == np.array(matrix1.matrix) - number).all()
    ), 'sub section with matrix and number is not working right'

@pytest.mark.parametrize("matrix1", [matrix1])    
def test_matrix_transpose(matrix1: Matrix):
    assert (
        (np.array(matrix1.T().matrix) == np.array(matrix1.matrix).T).all()
    ), 'transpose section is not working right'
    
# -----------------------------------------

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2]])
def test_matrix_mul_matrix(matrix1: Matrix, matrix2: Matrix):
    assert (
        (np.array((matrix1 * matrix2).matrix) == np.array(matrix1.matrix) * np.array(matrix2.matrix)).all()
    ), 'mul section with matrices is not working right'

#не работает
@pytest.mark.parametrize("matrix1,number", [[matrix1, 13]])
def test_matrix_mul_number(matrix1: Matrix, number: int | float):
    assert (
        (np.array((matrix1 * number).matrix) == np.array(matrix1.matrix) * number).all()
    ), 'mul section with matrix and number is not working right'

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2]])    
def test_matrix_div_matrix(matrix1: Matrix, matrix2: Matrix):
    assert (
        (np.array((matrix1 / matrix2).matrix) == np.array(matrix1.matrix) / np.array(matrix2.matrix)).all()
    ), 'div section with matrices is not working right'

#не работает
@pytest.mark.parametrize("matrix1,number", [[matrix1, 2]])    
def test_matrix_div_number(matrix1: Matrix, number: int | float):
    assert (
        (np.array((matrix1 / number).matrix) == np.array(matrix1.matrix) / number).all()
    ), 'div section with matrix and number is not working right'

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2.T()]])    
def test_matrix_matmul_matrix(matrix1: Matrix, matrix2: Matrix):
    assert(
        (np.array((matrix1 @ matrix2).matrix) == np.array(matrix1.matrix) @ np.array(matrix2.matrix)).all()
    ), 'matmul section is not working right, and don\'t forget about T() if needed'
```
